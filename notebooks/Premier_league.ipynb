```python
import pandas as pd
import ssl
import qgrid
from urllib import request


url="https://www.bbc.co.uk/sport/football/premier-league/table"
context = ssl._create_unverified_context()
response = request.urlopen(url, context=context)
html = response.read()

prem_table = pd.read_html(html)
Premier_table = prem_table[0]

print(len(prem_table))
print(type(prem_table))

Premier_table.head(18)
```

```python
Premier_table.columns
```

```python
Premier_league = Premier_table.drop(['Unnamed: 1'], axis=1)
```

```python
Premier_league.head()
```

```python
Premier_league.dtypes
```

```python
Premier_league[['F', 'A', 'P', 'W', 'D', 'L', 'GD', 'Pts']] = Premier_league[['F', 'A', 'P', 'W', 'D', 'L', 'GD', 'Pts']].apply(pd.to_numeric, errors='coerce')
```

```python
Premier_league.tail(3)
```

```python
Premier_league.drop(Premier_league.tail(1).index, inplace=True)
```

```python
Premier_league.rename(columns={'Unnamed: 0' : 'Position'}, inplace=True)
Premier_league.head()
```

```python
Premier_league['Goal Ratio'] = round(Premier_league['F'] / Premier_league['A'], 1)
Premier_league.head(3)
```

```python
Premier_league = Premier_league[['Position', 'Team', 'P', 'W', 'D', 'L', 'F', 'A', 'Goal Ratio', 'GD', 'Pts']]
Premier_league.head(3)
```

```python
one_less_fixture = Premier_league.P == 4
```
